```python
import pandas as pd
```

```python
data = pd.read_csv('Euro_2012_stats_TEAM.csv')
data
```

## **Which Team has the max Total shots (inc. Blocked)**

```python
data.loc[data["Total shots (inc. Blocked)"].idxmax(), "Team"]
```

## **Which team has the max Shooting Accuracy**

```python
data.loc[data["Shooting Accuracy"].map(lambda x: float(x[:-1])).idxmax()]
```

## **Teams with Penalty goals**

```python
data[data["Penalty goals"] > 0]
```

## **Mean Shooting Accuracy for teams with more than 5 Yellow Cards and less**

```python
print(data[data["Yellow Cards"] <= 5]["Shooting Accuracy"].map(lambda x: float(x[:-1])).mean())
data[data["Yellow Cards"] > 5]["Shooting Accuracy"].map(lambda x: float(x[:-1])).mean()
```

## **Percentage of blocked shots**

```python
data_tmp = data.copy()
data_tmp['Percent of blocked shots']  = (data['Total shots (inc. Blocked)'] - data['Shots on target'] - data['Shots off target']) * 100 / data['Total shots (inc. Blocked)']
data_tmp
```

## **Percentage of Yellow/Red Cards given for Fouls**

```python
data["Correspondence % between fouls and cards issued"] = round((data["Yellow Cards"] + data["Red Cards"]) * 100 /
                                                                (data["Fouls Won"] + data["Fouls Conceded"]), 2).astype(str) + "%"
data
```

## **Average value of Passes per game**

```python
data.Passes.mean()
```

## **Number of passes per game**

```python
data.Passes.sum()
```

## **Filter for teams that made > 2000 Passes completed and Shooting Accuracy below 30%**

```python
data["Shooting Accuracy"] = data["Shots on target"] * 100 /(data["Shots on target"] + data["Shots off target"])
data.loc[(data["Passes completed"] > 2000) & (data["Shooting Accuracy"] > 30)]
```

## **The average number of Touches to the ball when Dribbles**

```python
round((data["Touches"] - data["Passes"]) / data["Dribbles"], 2).mean()
```

## **The team with the best performance (the highest percentage between goals scored, passes, the fewest cards issued, and the fewest substitutions)**

```python
data["k"] = data["Goals"] * 0.4 + data["Passes completed"] * 0.1 + (data["Red Cards"] + data["Yellow Cards"]) * 0.3 + data["Subs on"] * 0.2
data.loc[data.k.idxmax()].Team
```

## **How many teams with no red card, with the Clearances more than 100 and Tackles less than 150**

```python
len(data.loc[(data["Red Cards"] == 0) & (data["Clearances"] > 100) & (data["Tackles"] < 150)])
```
