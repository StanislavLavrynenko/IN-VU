```python
def func():
    from random import randint
    """Write a program that lets the user play Rock-Paper-Scissors against the computer.
    There should be three rounds, and after those three rounds, your program should print out
    who won and lost or that there is a tie."""
    print(f" {'⏞⏞⏞'*14}\n \
              Let's play Rock-Paper-Scissors.\n \
{'⏟⏟⏟'*14}\n\n \
    Rules of the game:\n \
   ⌯ There are three rounds in the game\n \
   ⌯ Scissors win at Paper but lose at Rock\n \
   ⌯ Paper wins at Rock but loses at Scissors\n \
   ⌯ Rock wins at Scissors but loses at Paper\n\n \
{'⏛⏛⏛'*10}\n \
       Corresponds to Rock value -- 0 \n \
       Corresponds to Paper value -- 1 \n \
       Corresponds to Scissors value -- 2\n \
{'⏛⏛⏛'*10}\n ")
    game = True
    variables = ["Rock", "Paper", "Scissors"]
    pl_win = {0: 2, 1: 0, 2: 1}
    while game:
        pl = 0
        pc = 0
        for round in range(3):
            player_move = int(input(f"\n {'~~~'*6} Round {round + 1} {'~~~'*6} \n\n  Enter your move - "))
            computer_move = randint(0, 2)
            print(f"\n{'・・'*12} \n・      Your move - {variables[player_move]}\n・      \
Computer move - {variables[computer_move]}\n{'・・'*12}")
            if pl_win[player_move] == computer_move:
                pl += 1
                print(f"\n{'⌔⌔⌔'*12}\n⌔      In this round You win       ⌔\n{'⌔⌔⌔'*12}")
            elif player_move == computer_move:
                print(f"\n{'⌔⌔⌔'*12}\n⌔      In this round It's tie      ⌔\n{'⌔⌔⌔'*12}")
            else:
                pc += 1
                print(f"\n{'⌔⌔⌔'*12}\n⌔    In this round Computer win    ⌔\n{'⌔⌔⌔'*12}")
        print(f"{'---'*20}")
        if pl > pc:
            print(f"\n{'⏁⏁⏁'*8}\n⏀\n⏀     Congrutulations, You win\n⏀\n{'⏂⏂⏂'*8}")
        elif pl == pc:
            print(f"\n{'⏁⏁⏁'*8}\n⏀\n⏀     Tie, not bad, try one more time\n⏀\n{'⏂⏂⏂'*8}")
        else:
            print(f"\n{'⏁⏁⏁'*10}\n⏀\n⏀     You lose, try one more time, I think you can win me\n⏀\n{'⏂⏂⏂'*10}")
        cont = int(input("Do you want to try again?(0 - No/ 1- Yes): "))
        if cont == 1:
            print(f"\n{'⏛⏛⏛'*12}")
        else:
            game = False
        
                
                
func()
```
